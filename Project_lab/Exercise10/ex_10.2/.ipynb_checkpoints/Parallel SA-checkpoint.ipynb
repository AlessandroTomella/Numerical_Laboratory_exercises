## Parallel simulated annealing

Raccogliamo qui i risultati del codice in parallelo per la soluzione, tramite simulated annealing, del problema del commesso viaggiatore con 70 citt√† su un cerchio e poi nel quadrato. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def draw(array, x, y):
    
    array = array.astype(int)
    L = len(array)
    xpairs = []
    ypairs = []
    for k in range(L-1):
        xends = [x[array[k]], x[array[k+1]]]
        yends = [y[array[k]], y[array[k+1]]]
        xpairs.append(xends)
        ypairs.append(yends)
    xends = [x[array[L-1]], x[array[0]]]
    yends = [y[array[L-1]], y[array[0]]]
    xpairs.append(xends)
    ypairs.append(yends)
    plt.plot(xpairs, ypairs)
    
```

```python
f = plt.figure(figsize = (12, 6))

x, y = np.loadtxt("circle.start", usecols = (0,1), unpack = 'true')

for i in range(2):
    f.add_subplot(1, 2, i+1)
    plt.scatter(x, y, color = 'red')
    path = np.loadtxt("Circlepath"+str(i)+".dat")
    
    draw(path, x, y)    
```

```python
f = plt.figure(figsize = (12, 6))

x, y = np.loadtxt("square.start", usecols = (0,1), unpack = 'true')

for i in range(2):
    f.add_subplot(1, 2, i+1)
    plt.scatter(x, y, color = 'red')
    path = np.loadtxt("Squarepath"+str(i)+".dat")
    draw(path, x, y)    
```
