## Plot delle configurazioni finali di 3 simulazioni

### Configurazione iniziale fcc

```python
%matplotlib notebook
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x, y, z = np.loadtxt("config.fcc", usecols= (0, 1, 2),  unpack = True) 
d = len(x)
ax.plot3D(x, y, z, linewidth = 0.0, marker = '.')

plt.show()
```
