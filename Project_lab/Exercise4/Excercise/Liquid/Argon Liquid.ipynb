```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def error(AV1,AV2,n):  # Function for statistical uncertainty estimation
    if n==0:
        return 0
    else:
        return math.sqrt((AV1[n] - AV2[n])/n)
```

## Argon Liquid

```python
amu = 1.6605 * 10**(-27)   #kilograms
k_B = 1.38065 * 10**(-23)  #Joules/Kelvin
sigma = 0.34*10**(-9)      #meters
epsilon = 120 * k_B        #Joules
m = 39.948 * amu           #kilograms
tau = sigma*m**(1/2)*epsilon**(-1/2) #seconds
```

```python
# Potential energy
ep_ave, ep2_ave, ep_ave2 = np.loadtxt("ave_epot.dat", usecols = (0, 1, 2), unpack = 'true')
N = len(ep_ave)
n =np.arange(N)
e = np.zeros(N)

for i in range(N):
    e[i] = error(ep2_ave, ep_ave2, i)

plt.errorbar(n*tau*100, ep_ave*epsilon, e*epsilon) 
plt.xlabel('time [s]')
plt.ylabel('Potential Energy [J]')
plt.title('Potential energy/particle')
plt.grid(True)

print("Natural units:", ep_ave[N-1], "+/-", e[N-1])
print ("SI units:", ep_ave[N-1] * epsilon, "+/-" ,e[N-1] * epsilon, "J")
```

```python
# Kinetic energy
ek_ave, ek2_ave, ek_ave2 = np.loadtxt("ave_ekin.dat", usecols = (0, 1, 2), unpack = 'true')
N = len(ek_ave)
n =np.arange(N)
e = np.zeros(N)

for i in range(N):
    e[i] = error(ek2_ave, ek_ave2, i)

plt.errorbar(n*tau*100, ek_ave*epsilon, e*epsilon)
plt.xlabel('time [s]')
plt.ylabel('Kinetic Energy [J]')
plt.title('Kinetic energy/particle')
plt.grid(True)

print("Natural units:", ek_ave[N-1], "+/-", e[N-1])
print ("SI units:", ek_ave[N-1] * epsilon, "+/-" ,e[N-1] * epsilon, "J")
```

```python
# Total energy
e_ave, e2_ave, e_ave2 = np.loadtxt("ave_etot.dat", usecols = (0, 1, 2), unpack = 'true')
N = len(e_ave)
n =np.arange(N)
e = np.zeros(N)

for i in range(N):
    e[i] = error(e2_ave, e_ave2, i)

plt.errorbar(n*tau*100, e_ave*epsilon, e*epsilon)
plt.xlabel('time [s]')
plt.ylabel('Energy [J]')
plt.title('Total energy/particle')
plt.grid(True)

print("Natural units:", e_ave[N-1], "+/-", e[N-1])
print ("SI units:", e_ave[N-1] * epsilon, "+/-" ,e[N-1] * epsilon, "J")
```

```python
# Temperature
t_ave, t2_ave, t_ave2 = np.loadtxt("ave_temp.dat", usecols = (0, 1, 2), unpack = 'true')
N = len(t_ave)
n =np.arange(N)
e = np.zeros(N)

for i in range(N):
    e[i] = error(t2_ave, t_ave2, i)

plt.errorbar(n*tau*100, t_ave * epsilon/k_B, e * epsilon/k_B)
plt.xlabel('time [s]')
plt.ylabel('Temperature [K]')
plt.title('Mean Temperature')
plt.grid(True)

print("Natural units:", t_ave[N-1], "+/-", e[N-1])
print ("SI units:", t_ave[N-1] * epsilon/k_B, "+/-" ,e[N-1] * epsilon/k_B, "K")
```

```python
# Pressure
p_ave, p2_ave, p_ave2 = np.loadtxt("ave_press.dat", usecols = (0, 1, 2), unpack = 'true')
N = len(p_ave)
n =np.arange(N)
e = np.zeros(N)

for i in range(N):
    e[i] = error(p2_ave, p_ave2, i)

plt.errorbar(n*tau*100, p_ave*epsilon/sigma**3 , e*epsilon/sigma**3 )
plt.xlabel('time [s]')
plt.ylabel('Pressure [Pa]')
plt.title('Pressure')
plt.grid(True)

print("Natural units:", p_ave[N-1],'+/-', e[N-1])
print("SI units", p_ave[N-1] * epsilon/sigma**3,'+/-', e[N-1] * epsilon/sigma**3, "Pa")
```
