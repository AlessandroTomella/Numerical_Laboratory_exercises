```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

r, g_solid, err_g_solid = np.loadtxt("RDF(0).dat", usecols = (0, 1, 2), unpack = 'true')
plt.errorbar(r, g_solid, err_g_solid, label = 'Solid Phase')

r, g_liquid, err_g_liquid = np.loadtxt("RDF(1).dat", usecols = (0, 1, 2), unpack = 'true')
plt.errorbar(r, g_liquid, err_g_liquid, label = 'Liquid phase')

r, g_gas, err_g_gas = np.loadtxt("RDF(2).dat", usecols = (0, 1, 2), unpack = 'true')
#N = len(g_gas)
#r = np.linspace(0, 3, num = N)
plt.errorbar(r, g_gas, err_g_gas, label = 'Gaseous phase')
plt.xlim(0,3)
plt.xlabel('distance ($LJ$ units)')
plt.ylabel('g')
plt.title('Radial distribution function')
plt.grid()
plt.legend()
```
