## Modello di Ising 1D nel range di temperatura $[0.5, 2]$

Si è simulato un modello di Ising 1D con 50 gradi di libertà e con PBC tramite l'algoritmo di Metropolis. In particolare si è campionata la distribuzione di probabilità (Boltzmann) con i due metodi del $\textit{Metropolis standard sampling}$ e del $\textit{Gibbs sampling}$, per 100 diverse temperature all'interno del range $[0.5, 2]$.

Si sono quindi usati i risultati del campionamento per calcolare i valori medi e le incertezze, tramite il metodo del blocking, delle seguenti quantità termodinamiche, in funzione della temperatura:
- Energia interna per particella $U(T)$
- Capacità termica $C(T)$
- Magnetizzazione per particella $M(T)$
- Suscettibilità magnetica $\chi(T)$

È utile osservare che, per ogni temperatura, la simulazione consisteva in 

$$
N = 2\times10^5  \text{step MC}\hspace{0.5 cm}
\left\{
	\begin{array}{ll}
	N_{blocks} = 20	\\
    N_{step} = 10^4
	\end{array}
\right.
$$

dove uno "step MC" consisteva nel tentativo di "flippare" $\textit{tutti}$ e 50 gli spin. In questo modo si sono ridotte notevolmente le possibili correlazioni. Proprio per questo motivo è stato anche possibile osservare che la fase di equilibrazione del sistema era brevissima (qualche decina di MC steps), e si è quindi scelto di "incatenare" tutte le simulazioni e di registrare tutti i valori, anche quelli iniziali: in altre parole la simulazione del sistema a temperatura $T_i$ partiva dalla configurazione finale della simulazione a temperatura $T_{i-1}$, e il punto di partenza è stato la simulazione con $T = 2$, che è invece stata preceduta da una fase di equilibrazione esclusa dai risultati, essendo la configurazione iniziale una configurazione random corrispondente a $T = \infty$.

Vediamo quindi i risultati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
temp = np.loadtxt("output/MT_Energy.dat", usecols= (0))

# Carico i valori medi ottenuti con Metropolis standard sampling...
MT_obs = [ np.loadtxt("output/MT_Energy.dat", usecols= (2)),
  np.loadtxt("output/MT_Heat.dat", usecols= (2)),
  np.loadtxt("output/MT_Mag.dat", usecols= (2)),
  np.loadtxt("output/MT_Chi.dat", usecols= (2))]

# ... e i corrispondenti errori statistici
MT_err = [ np.loadtxt("output/MT_Energy.dat", usecols= (3)),
  np.loadtxt("output/MT_Heat.dat", usecols= (3)),
  np.loadtxt("output/MT_Mag.dat", usecols= (3)),
  np.loadtxt("output/MT_Chi.dat", usecols= (3))]

# Carico i valori medi ottenuti con il Gibbs sampling...
G_obs = [ np.loadtxt("output/Gibbs_Energy.dat", usecols= (2)),
  np.loadtxt("output/Gibbs_Heat.dat", usecols= (2)),
  np.loadtxt("output/Gibbs_Mag.dat", usecols= (2)),
  np.loadtxt("output/Gibbs_Chi.dat", usecols= (2))]

# ... e i corrispondenti errori statistici 
G_err = [ np.loadtxt("output/Gibbs_Energy.dat", usecols= (3)),
  np.loadtxt("output/Gibbs_Heat.dat", usecols= (3)),
  np.loadtxt("output/Gibbs_Mag.dat", usecols= (3)),
  np.loadtxt("output/Gibbs_Chi.dat", usecols= (3))]

# Nomi da inserire nei grafici
attr = ['Internal energy $U$ ($h = 0$)', 'Heat capacity $C$ ($h = 0$)',
        'Magnetisation $M$ ($h = 0.02$)', 'Magnetic susceptibility $\chi$ ($h = 0$)']

# Ricostruisco le curve teoriche esatte 
points=100
T = np.linspace(0.4,2.1,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th

# Energia interna
e = -J*( th + ch*thN )/( 1 + thN )
# Capacità termica
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

h=0.02 #external field
b = 1/T
l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
# Magnetizzazione
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

# Suscettibilità
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)

v = [e, heat, M, X]
f = plt.figure(figsize = (12, 10))

# In un ciclo plotto tutte le quantità
for i in range(4):
    graph = f.add_subplot(2, 2, i+1)
    graph.plot(T, v[i], label = 'Exact curve')
    graph.errorbar(temp, MT_obs[i], MT_err[i], label = 'Metropolis sampling')
    graph.errorbar(temp, G_obs[i], G_err[i], label = 'Gibbs sampling')
    plt.xlabel('temperature')
    plt.title(attr[i])
    plt.legend()
    plt.grid(True)
```

### Risultati

Osserviamo che nel range in analisi le simulazioni coincidono con le curve esatte in maniera soddisfacente. Si nota comunque un "peggioramento" man mano che la temperatura si avvicina allo $0$. Questo è dovuto alla combinazione di diversi fattori, che portano l'accettazione del Metropolis a essere molto bassa in queste condizioni e a peggiorare quindi le stime delle quantità medie in questa regione di temperatura.
